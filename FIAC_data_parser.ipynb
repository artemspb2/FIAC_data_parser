```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "http://www.feerc.obninsk.org:8080/Ru/Air2023w/air0423w.htm"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Поиск таблицы
table = soup.find('table')

# Получение всех строк таблицы
rows = table.find_all('tr')

# Создание списка для хранения данных из таблицы
data = []

# Перебор строк таблицы (за исключением заголовков)
for row in rows[2:]:
    # Получение всех ячеек в текущей строке
    cells = row.find_all('td')
    
    # Извлечение значений из ячеек и добавление их в список data
    values = [cell.get_text(strip=True) for cell in cells]
    data.append(values)

# Вывод данных
for row in data:
    print(row)
```

```python
# Создание датафрейма из списка данных
df = pd.DataFrame(data, columns=['N п/п', 'Название', 'Индекс', 'Широта, градусы', 'Долгота, градусы', 
                                 'Дата отбора пробы', 'Число дней до измерения', 'Максим. значение', 
                                 'Среднее значение', 'Число измерений', 'Дата отбора пробы', 
                                 'Число дней до измерения', 'Максим. значение', 'С/Б', 
                                 'Среднее значение', 'Число измерений'])
```
