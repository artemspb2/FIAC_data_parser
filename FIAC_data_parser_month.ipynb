```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen
```

```python
url = "http://www.feerc.obninsk.org:8080/Ru/Atm2023w/atm0123w.htm"
response = requests.get(url)
page = urlopen(url)
html = page.read().decode('windows-1251')

soup = BeautifulSoup(html, "html.parser")

# Поиск таблицы
table = soup.find('table')

# Получение всех строк таблицы
rows = table.find_all('tr')

# Создание списка для хранения данных из таблицы
data = []

# Перебор строк таблицы (за исключением заголовков)
for row in rows[2:]:
    # Получение всех ячеек в текущей строке
    cells = row.find_all('td')
    
    # Извлечение значений из ячеек и добавление их в список data
    values = [cell.get_text(strip=True) for cell in cells]
    data.append(values)

# Вывод данных
for row in data:
    print(row)
```

```python
# Создание датафрейма из списка данных
df = pd.DataFrame(data, columns=['N п/п', 'Название', 'Индекс', 'Широта', 'Долгота', 'Дата отбора пробы', 'Число дн. до измер.', 'Макс. знач.', 'Сред. знач.', 'Дата отбора пробы', 'Число дн. до измер.', 'Макс. знач.', 'С/Б', 'Сред. знач.'])
```
